```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

### 1. Cleaning Scraped Data

```python
df = pd.read_csv('datasets/glassdoor_jobs.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
kf = df['Salary Estimate'] == '-1'
df[kf].shape
```

seems like the null values are registered as negative ones, let's get rid of them

```python
df = df[~kf]
df.shape
```

```python
df.isnull().sum()
```

Removing '(Glassdoor est.)' from salary column

```python
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
df.head()
```

```python
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x: x.replace('K', '').replace('$', ''))
df.head()
```

#### Certain salary estimates seem to be in per hour rate, let's create a new feature that will encode whether this is true for a given entry

```python
df['hourly'] = df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
df.head()
```

#### same for the 'employer provied salary' remark

```python
df['employer'] = df['Salary Estimate'].apply(lambda x: 1 if 'employer provided salary'
                                             in x.lower() else 0)
df.head()
```

```python
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x: x.lower().replace('per hour', '')
                                  .replace('employer provided salary:', ''))
df['Salary Estimate'].unique()
```

```python
df['min_salary'] = df['Salary Estimate'].apply(lambda x: int(x.split('-')[0] + '000'))
df['max_salary'] = df['Salary Estimate'].apply(lambda x: int((x.split('-')[1]).strip() + '000'))
df.head()
```

```python
df = df.drop('Salary Estimate', axis=1)
```

```python
print(df.info())
df.head()
```

```python
df['avg_salary'] = (df['min_salary'] + df['max_salary']) / 2
df.head()
```

#### Salary column is done, now let's clean 'company name' so that the only value remaining is merely a company name

```python
df['Company Name'].unique()
```

entries that do not have rating (i.e -1) have just the right form of company name, otherwise the right form would be all characters in the string except last 4

```python
df['Company Name'] = df.apply(lambda x: x['Company Name'] if x['Rating']<0 else x['Company Name'][:-4], axis=1) 
df['Company Name'].unique()
```

##### I doubt we'll use company names for anything important, but at least they will be readable for summary statistics

let's now clean the location column and extract state variable from it

```python
df['state'] = df['Location'].apply(lambda x: (x.split(',')[1]).strip())
df.head()
```

##### How many jobs are in each state?


```python
df.state.value_counts()
```

##### Is the job position located at the headquarters?

```python
df['at_headquarters'] = df.apply(lambda x: 1 if x.Headquarters == x.Location else 0, axis=1)
```

```python
df.head()
```

##### Age of the company

```python
df['company_age'] = df.apply(lambda x: 2021 - x.Founded if (x.Founded>0) else x.Founded, axis=1)
df.head()
```

##### Let's now process the job description column to see which data science tools job requires. adding new binary columns for each tool would be one way to do it

```python
df['python_yn'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
df.python_yn.value_counts()
```

```python
#r studio 
df['R_yn'] = df['Job Description'].apply(lambda x: 1 if 'r studio' in x.lower() or 'r-studio' in x.lower() else 0)
df.R_yn.value_counts()
```

```python
#spark 
df['spark'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0)
df.spark.value_counts()
```

```python
#aws 
df['aws'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
df.aws.value_counts()
```

```python
#excel
df['excel'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
df.excel.value_counts()
```
